## Задача 2: аппроксимация функции

f(x) = sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)

```python
import numpy as np
from math import sin, exp
from scipy.linalg import solve
import matplotlib.pyplot as plt

def target_func(x):
    return sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)

f1 = target_func(1)
f15 = target_func(15)
A = np.array([[1,1], [1,15]])
b = np.array([f1, f15])
# print(A)
factors = solve(A, b)
def linear_func(x):
#     c = ax + b
    return factors[1] * x + factors[0]

fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.set_title("Linear approximation")  # Add a title to the axes.
ax.plot([x for x in range(0,17)], [target_func(x) for x in range(0,17)], label = 'original')
ax.plot(list([x for x in range(0,17)]), list([linear_func(x) for x in range(0,17)]), label = 'linear')
ax.plot([1, 15], [linear_func(x) for x in [1, 15]], 'rx')
[ax.annotate('x = ' + str(x), (x, linear_func(x))) for x in [1, 15]]
# ax.set_xlabel('x')  # Add an x-label to the axes.
# ax.set_ylabel('y')  # Add a y-label to the axes.
ax.legend()  # Add a legend.

f4 = target_func(4)
f10 = target_func(10)

A = np.array([[1, 1, 1, 1],[1, 4, 16, 64], [1, 10, 100, 1000], [1, 15, 225, 225 * 15]])
b = np.array([f1, f4, f10, f15])
factors = solve(A, b)
def cubic_func(x):
    return factors[3] * x ** 3 + factors[2] * x ** 2 + factors[1] * x + factors[0]

fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.set_title("Cubic approximation")  # Add a title to the axes.
ax.plot([x for x in range(0,17)], [target_func(x) for x in range(0,17)], label = 'original')
ax.plot([1, 4, 10, 15], [cubic_func(x) for x in [1, 4, 10, 15]], 'rx')
ax.plot(list([x for x in range(0,17)]), list([cubic_func(x) for x in range(0,17)]), label = 'cubic')
[ax.annotate('x = ' + str(x), (x, cubic_func(x))) for x in [1, 4, 10, 15]]
ax.legend()  # Add a legend.
```
